What are mappings ?

Mappings are a collection of objects that are stored by a key, unlike a sequence that stored objects by their relative position. This is an important distinstion, since mapping won't retain order since they have objetcs by a key.

* A Python dictionary consists of a key and then an associated value. That value can be almost any Python object.


**Constructing a Dictionary**



```python
my_dict = {'key1':'value1','key2':'value2'}
my_dict['key1']
```

Dictionaries are very flexible in the data types they can hold.

```python
my_dict = {'key1':123,'key2':[12,23,33],'key3':['item0','item1','item2']}
my_dict['key3']
my_dict['key3'][2]
my_dict['key3'][2].upper()
```

We can affect the values of a key as well.

```python
my_dict['key1']
my_dict['key1'] = my_dict['key1'] - 123
my_dict['key1']
```

We can also create keys by assignment. For instance if we started off with an empty dictionary, we could continually add to it.

```python
d = {}
d['animal'] = 'dog'
d['num'] = 111
d[2] = 'name'
d[''] = 0
d
```

**Nesting With Dictionaries**

Python is with its flexibility of nesting objects.

```python
nested_dict = {'key1':{'nestkey':{'subnestkey':'value'}}}
nested_dict['key1']['nestkey']['subnestkey']
```

**A few Dictionary Methods**


```python
my_dict = {'key1':123,'key2':[12,23,33],'key3':['item0','item1','item2']}

my_dict.keys()
my_dict.values()
my_dict.items()
```

```python
grades = {
    'Alice': {'math' : 95, 'science': 90},
    'Bob': {'math' : 85, 'science': 80},
    'Carol': {'math' : 90, 'science': 95}
}

grades['Charlie'] = {'math': 89, 'science': 90}

for student, score in grades.items():
  total = sum(score.values())
  avg = total / len(score)
  print(f"{student}: {avg}")
```
