**Files**

Python uses file objects to interact withexternal files on your computer. These file objects can be any sort of file you have on your computer. whether it be an audio file, a text file, emails, Excel documents etc.

**IPython Writing a File**

```python
%%writefile test.txt
Hello,

Testing file writing.
```

```python
# Open the .txt file
openFile = open('/content/test.txt')

# Read the file
openFile.read()
```

This happens because you can imagine the reading "cursor" is at the end of the file after having read it. So there is nothing left to read.

We can rest the "cursor" like this

```python
openFile.read() # output = ''
```

```python
# We can rest the "cursor" like this
openFile.seek(0)
openFile.read()
```

You can read a file line by line using the readlines method. Use caution with large files, since everything will be held in memory

```python
# Readlines returns a list of the lines in the file
openFile.seek(0)
openFile.readlines()
```

```python
file_length = openFile.seek(0)
file_length = openFile.readlines()
print(len(file_length))
```

```python
openFile.close()
```

```python
# You can't read a closed file, to read you have open first
openFile.read()
```

**Writing to a File**

By default the ".open()" function will only allow us to read the file. We need to pass the argument 'w' to write over the file.

```python
myFile = open('/content/test.txt', 'w+')
```

Opening a file with 'w' or 'w+' truncates the original, meaning that anything that was in the original file *is deleted!*

```python
myFile.write('This is the new context')
# The output numbers refers to how many bits it takes
```

```python
myFile.seek(0)
myFile.read()
```

```python
myFile.close()
```

**Appending to a File**

Passing the argument `'a'` opens the file and puts the pointer at the end, so anything written is appended like `'w+'`, `'a+'` lets us read and write to a file. If the file doesn't exist, one will be created.

```python
myFile = open('/content/test.txt', 'a+')
myFile.write('\nThis is text being appended to test.txt')
myFile.write('\nAnd another line here.')
```

```python
myFile.seek(0)
print(myFile.read())
```

```python
myFile.close()
```
