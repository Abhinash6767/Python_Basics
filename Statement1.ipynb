**Introduction to Python Statements**

**if, elif, else Statements**


```python
if case1:
  statement1
elif case2:
  statement2
else:
  statement3
```

```python
age = 9
if age >= 18:
  print('You are allowed')
else:
  print('Not allowed')
```

```python
loc = 'Bank'
if loc == 'Auto Shop':
  print('Welcome to Auto Shop')
elif loc == 'Bank':
  print('Welcome to Bank')
elif loc == 'Mall':
  print('Welcome to mall')
else:
  print('Where are you?')
```
