**For Loops**

A `for` loop acts as an iterator in Python, it goes through items that are in a sequence or any other iterable item. We can iterate over include `Strings, Lists, Tuples` and even built-in iterables for `Dictionaries`, such as keys or values.

```
for item in object:
  statements to do stuff
```

````python
list1 = [1,2,3,'e',4,5, 'h']

for item in list1:
  if type(item) == str:
    print(item)
    # continue
  print(item)
````

````python
list2 = [1,2,3,4,3,4,5,3,4,5,6,3,23,4,88]

for number in list2:
  if number % 2 == 0:
    print(number)
  else:
    print('odd')
````

````python
total_sum = 0

for number in list2:
  total_sum += number
print(total_sum)
````

````python
hello = 'hello world'

for letter in hello:
  print(letter)
````

````python
tup = (1,2,3,4,5,'a', True)

for item in tup:
  print(item)
````

**Tuple Unpacking**

Tuples have a special quality when it comes to for loops. If you are iterating through a sequence that contains tuples, the item can actually be the tuple itself. Below is the example of tuple unpacking. During the for loop we will be unpacking the tuple inside of a sequence and we can access the individual items inside that tuple.

````python
list3 = [(1,2), (4,5), (8,9)]

for tup in list3:
  print(tup)
````

````python
# with unpacking
# One way

for (a,b) in list3:
  print(a)
  print(b)
````

````python
# with unpacking
# Another way

for (a) in list3:
  print(*a)
````

````python
list3 = [(1,2), (4,5), (8,9)]

for s, item in enumerate(list3):
  print(item)
````

````python
for s, item in enumerate(list3):
  print(item, end='--' if s < len(list3) -1 else '\n')
````

````python
list4 = [1,2,3,4,3,4,23,4,5,3,54,3,45,3,4,3,4,9]

sum = 0
for item in list4:
  if(item % 2 == 0):
    sum += item

print(sum)
````
