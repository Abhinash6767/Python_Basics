**Tuples**

In python tuples are very similar to lists, however, unlike *lists* they are *immutable* meaning they can not be changed. You would use tuples to present things that shouldn't be changed, such as days of the week, or dates on a calender.

**Constructing Tuples**

The constuction of a tuples use "()" with elements separated by commas.



```python
t = ('abhi',8,9)
t
type(t)
```

```python
len(t)
```

```python
# Use indexing
t[0]
```

```python
# Slicing just like a list
t[-1]
```

**Basic Tuple Methods**

Tuples have built in methods, but not as many as lists do.

```python
# Use ".index" to enter a value and return the index
t.index(9)
```

```python
# Use ".count" to count the number of times a value appears
t.count(9)
```

**Immutability**

It can't be stressed enough that tuples are immutable.

```python
t[0] = "madhu"
```

```python
t.append("madhu")
```

```python
locations = {}
# Lists can't be used as a key in a Dictionary
locations[[23.345, 55.789]] = "India"
locations
```

```python

# Tuples can be used as a key in a Dictionary
locations[(23.345, 55.789)] = 'India'
locations
```

**When to use Tuples ?**

Tuples are not used as often as lists in programming but are used when immutability is necessary.  If in your program you are passing around an object and need to make sure it does not get chnaged, then a tuple becames your solution. It provides a convenient source of *Data Integrity*.
