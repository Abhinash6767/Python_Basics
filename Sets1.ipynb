**Sets and Booleans**

There are two other object types in Python, i.e Sets and Booleans.

**Sets**

Sets are an unordered collection of *Unique* elements. We can construct them by using the *set()* function.

```python
x = set()
```

```python
# We can add to sets with ".add" method
x.add(1)
```

```python
list1 = [1,4,5,7,9,6,51,1,4,5,9]
list1.sort()
list1
```

```python
set(list1)
```

```python
my_set = {6,9,8,432,54,7}
my_sortedSet = sorted(my_set)
my_sortedSet
```

**Booleans**

Python comes with Booleans (with predefined True and False that are basically just the integers 1 and 0). It also has a *placeholder object called None*.

```python
a = True
a
```

```python
1 == 2
12 > 10
```

We can use None as a placeholder for an object that we don't want to reassign yet.

```python
b = None
print(b)
```
